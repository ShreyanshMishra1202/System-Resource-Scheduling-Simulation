```python
drones=[{"id":1,"angle":5,"speed":5,"visible":False,"priority":4},
       {"id":2,"angle":0,"speed":7,"visible":False,"priority":2}]
```

```python
tower={"busy":False,
       "current_drone":None
      }
```

```python
time_slot = 0
dr1=None

for time in range(0, 1440):
    print("Time:", time)

    avl=[]
    
    # Move drone
    for drone in drones:
        drone["angle"] = (drone["angle"] + drone["speed"]) % 360

        # Check visibility (FIXED condition)
        if 30 <= drone["angle"] <= 150:
            drone["visible"] = True
            avl.append(drone)
        else:
            drone["visible"] = False

    avl=sorted(avl, key=lambda x: (x["priority"],x["id"]))


    if time_slot == 0:
        if len(avl)==0:
            print("Tower is free but no drone is visible")
        else:
            dr1=avl[0]
    

    

    # Tower decision
    if not tower["busy"] and dr1!=None:
        time_slot = 1
        tower["busy"] = True
        tower["current_drone"] = dr1["id"]
        print("Drone", tower["current_drone"], "started communication")
        

    else:
        if tower["busy"]:
            time_slot += 1
            if time_slot == 5:
                print("Drone", tower["current_drone"], "communication ended")
                tower["busy"] = False
                tower["current_drone"] = None
                time_slot = 0
                dr1=None
        
    if time_slot!=0:
        print("Angle",dr1["angle"],"and Visibility is",dr1["visible"])
```
