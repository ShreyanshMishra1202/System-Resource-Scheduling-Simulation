```python
drones=[{"id":1,"angle":5,"speed":5,"visible":False},
       {"id":2,"angle":0,"speed":7,"visible":False}]      ]
```

```python
tower={"busy":False,
       "current_drone":None
      }
```

```python
time_slot = 0

for time in range(0, 1440):
    print("Time:", time)

    # Move drone
    for drone in drones:
        drone["angle"] = (drone["angle"] + drone["speed"]) % 360

    # Check visibility (FIXED condition)
    if 30 <= drone["angle"] <= 150:
        drone["visible"] = True
    else:
        drone["visible"] = False

    # Tower decision
    if not tower["busy"]:
        if drone["visible"]:
            time_slot = 1
            tower["busy"] = True
            tower["current_drone"] = drone["id"]
            print("Drone", tower["current_drone"], "started communication")
        else:
            print("Tower is free but drone is not visible")

    else:
        time_slot += 1
        if time_slot == 5:
            print("Drone", tower["current_drone"], "communication ended")
            tower["busy"] = False
            tower["current_drone"] = None
            time_slot = 0
    print("Angle",drone["angle"],"and Visibility is",drone["visible"])
```
